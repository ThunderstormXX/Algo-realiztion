```python
import cv2 as cv
import numpy as np
from matplotlib import pylab as plt
```

```python
def comap(image_first , func):
    image = image_first.copy()
    X = image.shape[0]
    Y = image.shape[1]
    image = image*0
    print(X,Y)
    for i in range(X):
        for k in range(Y):
            point = np.complex(i + k*1j) 
            point = func(point,X,Y)
            x = int(np.real(point)) 
            y = int(np.imag(point))
            if x < X and y < Y and 0 <= x and 0<= y :
                image[x][y] = image_first[i][k]
    plt.imshow(image)
    return image
```
