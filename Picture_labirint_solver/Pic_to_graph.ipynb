```python
import cv2 as cv
import numpy as np
from matplotlib import pylab as plt

image = cv.imread("test_img_1.png")
```

```python
def coloring(image , k , color):
    j = k % image.shape[1]
    i = k // image.shape[0]
    print(i,j)
    image[i][j] = color
```

```python
def Search_path(image_first , x_1 , x_2 ):
     image = image_first.copy()
     X = image.shape[0]
     Y = image.shape[1]
     s = x_1[0]*Y +x_1[1]
     t = x_2[0]*Y + x_2[1]
     n = X*Y
     G =[[-1 for i in range(n)] for i  in range(n)]
     for i in range(X):
          for j in range(Y):
               if i + 1 < X and tuple(image[i+1][j]) == (255,255,255):
                    G[Y*(i)+j][Y*(i+1)+j] = 1
               if i - 1 > -1 and tuple(image[i-1][j]) == (255,255,255):
                    G[Y*(i)+j][Y*(i-1)+j] = 1
               if j + 1 < Y and tuple(image[i][j+1]) == (255,255,255):
                    G[Y*(i)+j][Y*(i)+j+1] = 1
               if j - 1 > -1 and tuple(image[i][j-1]) == (255,255,255):
                    G[Y*(i)+j][Y*(i)+j-1] = 1
     # for i in G:
     #      print(i)

     visited = []
     queue = [] 
     d = [-1 for i in range(n)]  
     visited.append(s) 
     queue.append(s)
     d[s] = 0
     p = [-1 for i in range(n)]
     while queue:
          m = queue.pop(0)
          for v in range(n):
               if (G[m][v] == 1) & (v not in visited):
                    visited.append(v)
                    queue.append(v)
                    d[v] = d[m] + 1
                    p[v] = m
     c = t
     #print(p)
     while p[c] != -1 :
          #print(c, s , p[c])
          #coloring(image,c,(100,200,100))
          c = p[c]
     for i in p:
          if i < X*Y :
               print(i)
     
     
     coloring(image,s,(50,70,240))
     coloring(image,t,(5,7,240))
     plt.imshow(image)
```
